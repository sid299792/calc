```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np

style.use("ggplot")
```

```python
fn = lambda x:1 + (1/2)*(x**2)
start = -50
stop = 50
step = 0.01
```

```python
def draw(fn,start,stop,step):
    count = start
    x = []
    y = []
    print(f"This graph will contain {(stop-start)//step} points")
    while count <= stop:
        x.append(count)
        y.append(fn(count))
        count += step
    plt.plot(x,y)
```

```python
draw(fn,start,stop,step)
```
