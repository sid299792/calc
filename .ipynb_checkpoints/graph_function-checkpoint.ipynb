```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import math

style.use("ggplot")
```

```python
fn = lambda x:1 - (1/2)*(x**2)
start = math.radians(-180)
stop = math.radians(180)
step = math.radians(0.1)
```

```python
def draw(fn,start,stop,step):
    count = start
    x = []
    y = []
    print(f"This graph will contain {(stop-start)//step} points")
    while count <= stop:
        x.append(count)
        y.append(fn(count))
        count += step
    plt.plot(x,y)
```

```python
draw(lambda x:math.fabs(fn(x)-math.cos(x)),start,stop,step)
plt.show()
```
